```python
# Dependencies and Setup
import pandas as pd

```

```python
# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data.head (5)
```

```python
purchase_data.dtypes
```

Player Count

```python
# Total Number of Players
players = purchase_data.groupby("SN")["SN"].nunique()
players.count()
players = players.count()

pd.DataFrame({"Total Players":[players]})
```

# Purchasing Analysis (Total)

```python
# Number of Unique Items
items = purchase_data.groupby("Item ID")["Item ID"].nunique()
items = items.count()

# Average Purchase Price
price = purchase_data["Price"].mean()
# price

# Total Number of Purchases
numbers = purchase_data["Price"].count()
# numbers

# Total Revenue
revenue = purchase_data["Price"].sum()
# revenue

purchasing_analysis= pd.DataFrame({"Number of Unique Items":[items],
                                 "Average Price":[price],
                                  "Number of Purchases":[numbers],
                                    "Total Revenue":[revenue]
                                })

purchasing_analysis.style.format({ "Average Price":"${:,.2f}", 
                                "Total Revenue":"${:,.2f}",
                                })
```

# Gender Demographics

```python
# Percentage and Count of Male Players, # Percentage and Count of Female Players, 
# Percentage and Count of Other / Non-Disclosed

gender_count = purchase_data.groupby("Gender")["SN"].nunique()

gender_percents = (gender_count/players*100).round(2).astype(str) +'%'

gender_demographics = pd.DataFrame({"Total Count":gender_count,
             "Percentage of Players":gender_percents
            })

# map('{:.2f}%'.format), # astype(int).astype(str).add('%'), # round(2) --> these functions are useful for formating

gender_demographics
```

# Purchasing Analysis (Gender)

```python
# The below each broken by gender, # Purchase Count

purchase_count = purchase_data.groupby("Gender")["Item Name"]
purchase_count.count()

# Average Purchase Price,
average_price = purchase_data.groupby("Gender")["Price"].mean()
average_price.round(2)

# Total Purchase Value
total_value_gender = purchase_data.groupby("Gender")["Price"].sum()
total_value_gender

# Average Purchase Total per Person by Gender

```

```python
# average_purchase_person_gender
average_purchase_person_gender = (total_value_gender/gender_count)
average_purchase_person_gender.round(2)
```

```python
# Purchasing Analysis (Gender) summary
purchase_summary = pd.DataFrame({"Purchase Count":purchase_count.count(),
                                 "Average Purchase Price":average_price.round(2),
                                 "Total Purchase Value":total_value_gender,
                                 "Avg Total Purchase per Person":average_purchase_person_gender,
                                })
purchase_summary

purchase_summary.style.format({ "Average Purchase Price":"${:,.2f}", 
                                "Total Purchase Value":"${:,.2f}",
                                "Avg Total Purchase per Person":"${:,.2f}"
                                })
```

# Age Demographics

```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Segment and sort age values into bins established above
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels=group_names)
# purchase_data
# Create new data frame with the added "Age Group" and group it
age_grouped = purchase_data.groupby("Age Group")
# Count total players by age category
total_player_age = age_grouped["SN"].nunique()
total_player_age
# Calculate the numbers and percentages by age group
age_group_percent = (total_player_age/players)*100
age_group_percent = age_group_percent.round(2).astype(str).add('%')


# Create a summary data frame to hold the results
age_demographics = pd.DataFrame({"Total Count":total_player_age,
             "Percentage of Players":age_group_percent
            })
age_demographics
# Optional: round the percentage column to two decimal points, # Display Age Demographics Table
```

# Purchasing Analysis (Age)

```python
# Bin the purchase_data data frame by age
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
# Run basic calculations to obtain purchase count age
purchase_count_age = age_grouped["Purchase ID"].count()
purchase_count_age

# calculate avg. purchase price by age
avg_purchase_price_age = age_grouped["Price"].mean()
avg_purchase_price_age.round(2)

# Calculate total purchase value by age group 
total_purchase_value = age_grouped["Price"].sum()
total_purchase_value

# Calculate the average purchase per person in the age group 
avg_purchase_per_person_age = total_purchase_value/purchase_count_age
avg_purchase_per_person_age.round(2)

# Create a summary data frame to hold the results

purchasing_analysis_age = pd.DataFrame({"Purchase Count":purchase_count_age,
                                "Average Purchase Price":avg_purchase_price_age.round(2),
                                "Total Purchase Value":total_purchase_value,
                                "Avg Total Purchase per Person":avg_purchase_per_person_age.round(2),
                                       })
purchasing_analysis_age

purchasing_analysis_age.style.format({ "Average Purchase Price":"${:,.2f}", 
                                         "Total Purchase Value":"${:,.2f}",
                                        "Avg Total Purchase per Person":"${:,.2f}"
                                               })

# Optional: give the displayed data cleaner formatting, # Display the summary data frame
```

# Top Spenders

```python
# Run basic calculations to obtain the results in the table below
# Create a summary data frame to hold the results
gender_spenders = purchase_data.groupby("SN").count()["Price"]
gender_spenders

# Average Purchase Price by Spenders
average_purchase_price = purchase_data.groupby("SN").mean()["Price"]
average_purchase_price.round(2)
# average_purchase_price.sort_values(ascending=False)

# Total Purchase Value by Spenders
total_purchase_value = purchase_data.groupby("SN").sum()["Price"]
total_purchase_value

top_spenders= pd.DataFrame({"Purchase Count":gender_spenders,
                             "Average Purchase Price":average_purchase_price.round(2),
                             "Total Purchase Value":total_purchase_value,
                            })

top_spenders = top_spenders.sort_values(['Total Purchase Value'],ascending=False)

top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map('${:.2f}'.format)

top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map('${:.2f}'.format)

top_spenders.head()

# formated_top_spenders = top_spenders.style.format({ "Average Purchase Price":"${:,.2f}", 
#                             "Total Purchase Value":"${:,.2f}"
#                                 })
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame
```

# Most Popular Items

```python
# Retrieve the Item ID, Item Name, and Item Price columns
items = purchase_data[['Item ID','Item Name','Price']]
items

# Group the item data by item id and item name 
item_name = items.groupby(["Item ID","Item Name"])

# Group by Item ID and Item Name. 
# Perform calculations to obtain purchase count, item price, and total purchase value
items_purchase_count= item_name["Price"].count()
items_purchase_count

items_average_price = item_name["Price"].mean()
items_average_price

items_value_total = item_name["Price"].sum()
items_value_total

popular_items = pd.DataFrame({"Purchase Count":items_purchase_count,
                                "Average Purchase Price":items_average_price.round(2),
                                "Total Purchase Value":items_value_total.round(2),  
                           })

popular_items = popular_items.sort_values(['Purchase Count'],ascending=False)

popular_items["Average Purchase Price"] = popular_items["Average Purchase Price"].map('${:.2f}'.format)
popular_items["Total Purchase Value"] = popular_items["Total Purchase Value"].map('${:.2f}'.format)

popular_items.head()

# there are few ways to get this format value here there are.
# popular_formated = popular_items.sort_values(['Total Purchase Value'],ascending=False)
# formated_polular_items = popular_formated.style.format({ "Average Purchase Price":"${:,.2f}", 
#                                   "Total Purchase Value":"${:,.2f}"
#                                   })
```

```python
# Sort the above table by total purchase value in descending order

popular_items = pd.DataFrame({"Purchase Count":items_purchase_count,
                                "Average Purchase Price":items_average_price.round(2),
                                "Total Purchase Value":items_value_total.round(2),  
                           })

profitable_items = popular_items.sort_values(['Total Purchase Value'],ascending=False)

profitable_items["Average Purchase Price"] = profitable_items["Average Purchase Price"].map('${:.2f}'.format)
profitable_items["Total Purchase Value"] = profitable_items["Total Purchase Value"].map('${:.2f}'.format)

profitable_items.head()

# Optional: give the displayed data cleaner formatting
# Display a preview of the data frame
```
